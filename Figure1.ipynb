```python
from os.path import join
import numpy as np
from flyeye.data.experiments import Experiment
from figures.dynamics import Expression    
%matplotlib inline

data_path = '../data/published'
```

# Load wildtype data

```python
# define path to wildtype data
path = join(data_path, 'wildtype')

# load data
wildtype = Experiment(path, 0, align_by='ch2_normalized')

# add P/Y ratio to cell data
wildtype.set_ratio('ch1_normalized', 'ch2_normalized')
```

# Figure 1 H,I

```python
# define axis limits
ax_kw = dict(xlim=(-15, 40), 
             ylim=(.3, 1.2), 
             yticks=np.arange(.4, 1.3, .2))

# define number of bootstrap samples (10k used for manuscript)
interval_kw = dict(nbootstraps=10) 

# Fig 1H
fig1H = Expression(wildtype)
group1 = [['r8'], ['r2', 'r5'], ['r3', 'r4']]
fig1H.render('ch1_normalized', group1, scatter=False, interval=True, ax_kw=ax_kw, interval_kw=interval_kw)

# Fig 1I
fig1I = Expression(wildtype)
group2 = [['r1', 'r6'], ['r7']]
fig1I.render('ch1_normalized', group2, scatter=False, interval=True, ax_kw=ax_kw, interval_kw=interval_kw)
```
