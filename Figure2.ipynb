```python
import numpy as np
from os.path import join

from flyeye.data.experiments import Experiment
from figures.dynamics import Expression    
from figures.comparison import VerticalBoxplot
from figures.scatter import DualJointDistribution

%matplotlib inline

data_path = '../data/published'
```

# Load wildtype data

```python
# define path to wildtype data
path = join(data_path, 'wildtype')

# load data
wildtype = Experiment(path, 0, align_by='ch2_normalized')

# add P/Y ratio to cell data
wildtype.set_ratio('ch1_normalized', 'ch2_normalized')
```

# Figure 2 A-F

```python
# Fig 2ACE
fig = Expression(wildtype)
r2r5 = [['r2', 'r5']]
fig.render_all_channels(r2r5, scatter=True, shading='orange')

# Fig 2BDF
fig = Expression(wildtype)
r1r6 = [['r1', 'r6']]
fig.render_all_channels(r1r6, scatter=True, shading='orange')
```

# Figure 2 G

```python
fig = VerticalBoxplot.from_experiment(wildtype)
fig.render()
```

# Figure 2 H,I

```python
fig = DualJointDistribution.from_experiment(wildtype, ['r2', 'r5'])
fig.render()

fig = DualJointDistribution.from_experiment(wildtype, ['r1', 'r6'])
fig.render()
```
