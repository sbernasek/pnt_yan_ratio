```python
from os.path import join
import numpy as np    

from flyeye.data.experiments import Experiment

from figures.dynamics import DualExpression
from figures.comparison import PerturbationComparison

%matplotlib inline

data_path = '../data/published'
```

# Load EGFR ts experiment data

```python
EGFRts_control = Experiment(join(data_path, 'egfr_ts', '18c'), normalization='ch2', recompile=True)
EGFRts_perturbation = Experiment(join(data_path, 'egfr_ts', '26c'), normalization='ch2', recompile=True)

# evaluate P/Y ratio
EGFRts_control.set_ratio('ch1_normalized', 'ch0_normalized')
EGFRts_perturbation.set_ratio('ch1_normalized', 'ch0_normalized')
```

# Figure S1

```python
# define number of bootstrap samples (10k used for manuscript)
interval_kw = dict(nbootstraps=10) 

# Fig S8A (Pnt levels)
fig = DualExpression(EGFRts_control, EGFRts_perturbation)
ax_kw = dict(xlim=(-10, 30), xticks=np.arange(-10, 35, 10), 
             ylim=(0, 2), yticks=np.arange(0, 2.1, .5))
fig.render('ch1_normalized', interval=True, ax_kw=ax_kw, interval_kw=interval_kw)

# Fig S8B (Yan levels)
fig = DualExpression(EGFRts_control, EGFRts_perturbation)
ax_kw = dict(xlim=(-10, 30), xticks=np.arange(-10, 35, 10), 
             ylim=(0, 2.2), yticks=np.arange(0, 2.1, .5))
fig.render('ch0_normalized', interval=True, ax_kw=ax_kw, interval_kw=interval_kw)

# Fig S8C (Ratio)
fig = DualExpression(EGFRts_control, EGFRts_perturbation)
ax_kw = dict(xlim=(-10, 30), xticks=np.arange(-10, 35, 10), 
             ylim=(-1.4, 0.6), yticks=np.arange(-1.4, .64, .4))
fig.render('logratio', interval=True, ax_kw=ax_kw, interval_kw=interval_kw)
```
